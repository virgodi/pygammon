```python
from bs4 import BeautifulSoup
import requests
import time
```

# Matches from the Norges Backgammonforbund
### http://www.paulspages.co.uk/bgvaults/nbgf/index.php

```python
req = requests.get('http://www.paulspages.co.uk/bgvaults/nbgf/index.php')
page = req.text
soup = BeautifulSoup(page, 'html.parser')
```

```python
hrefs = []
for a in soup.find_all('a'):
    if a.get_text() == 'Download':
        hrefs.append(a['href'])
```

```python
norges_matches = {}
for h in hrefs:
    req = requests.get('http://www.paulspages.co.uk/bgvaults/nbgf/'+h)
    page = req.text
    norges_matches[h] = page
```

# Saving OG Games

```python
import pickle
```

```python
matches_dict = [norges_matches, onehundred2pts_matches, onehundred5pts_matches, twohundred7pts_matches, gnuvsblitz_matches]
```

```python
matches_name = ['norges', '100_3_pts', '100_5_pts', '200_7_pts', 'gnu_vs_blitz']
```

```python
for m, n in zip(matches_dict, matches_name):
    with open('og_matches/'+ n + '.pkl', 'w') as f:
        pickle.dump(m, f)
        
# with open('og_matches/norges.pkl', 'r') as f:
#     test = pickle.load(f)
```
